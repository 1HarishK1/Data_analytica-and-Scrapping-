```python
Titles=[]
```

```python
def find_title(url):
  try:
    response=requests.get(url)
    soup=BeautifulSoup (response.text,'html.parser')
    so=soup.find('div',class_='td-post-header')
    if so is None:
      title=soup.find('h1').text
      return title
    else:
      Head=so.find('header',class_='td-post-title')
      title=Head.find('h1',class_='entry-title').text
      return title
  except AttributeError:
    title=url.split('/')
    return title[3]
```

```python
for url in lst_urls:
    Title=find_title(url)
    Titles.append(Title)
```

```python
def preprocessing(headlines):
   clean_head=re.sub(r"[^a-zA-Z]",' ',headlines)
   clean_head=clean_head.lower()
   return clean_head.rstrip()
```

```python
for headline in Titles:
   cl_data=preprocessing(headline)
   Top.append(cl_data)
```

```python
stopwords=[]
import csv
```

```python
file=['StopWords_Auditor.txt',
      'StopWords_Generic.txt',
      'StopWords_DatesandNumbers.txt',
      'StopWords_Names.txt',
      'StopWords_Geographic.txt']
contents=[]
```

```python
for words in file:
   with open(words,'r') as words_data:
        conten=csv.reader(words_data)
        for i in conten:
          content=(i[0].lower())
          contents.append(content)
```

```python
def remove_words(headline):
   con=""
   H=(headline.split(' '))
   count=0
   for i in H:
     if i not in contents:
       con+=''.join(' '+ i)
       count+=1
   return con,count
```

```python
no_of_sent=[]
removed_words=[]
```

```python

for topic in Top:
   filter_sent=remove_words(topic)
   removed_words.append(filter_sent[0])
   no_of_sent.append(int(filter_sent[1]))
```

```python
posdata=[]
negdata=[]
```

```python
with open ('positive-words.txt','r') as data:
     words=csv.reader(data)
     for word in words:
       word=(word[0].lower())
       posdata.append(word)
```

```python
with open ('negative-words.txt','r', encoding="ISO-8859-1") as data:
     words=csv.reader(data)
     for word in words:
       negdata.append(word[0].lower())
```

```python
POSITIVE_SCORE=[]
NEGATIVE_SCORE=[]
```

```python
def score(sentence,len_word):
  splt=(sentence.split(' '))
  positivescore=len_word
  negativescore=-len_word
  for word in splt:
     if word in  posdata:
        positivescore+=1
     if word in negdata:
        negativescore*=-1
     else:
        pass
  return positivescore,negativescore
```

```python
for words in range(0,len(removed_words)):
    scor=score(removed_words[words],no_of_sent[words])
    POSITIVE_SCORE.append(int(scor[0]))
    NEGATIVE_SCORE.append(int(scor[1]))
```

```python
Polarityscore=[]
Subjectivescore=[]
```

```python
def polarscore(pos,neg,len_word):
  x=(pos-neg)
  y=(pos+neg)
  if x!=0 and y!=0:
    Pola=round((x/y),2)
    Subjective=round((y/len_word),2)
    return Pola,Subjective
  elif y!=0:
    Pola=0
    Subjective=round((y/len_word),2)
    return Pola,Subjective
  elif x==0 or y==0:
    Pola=0
    Subjective=0
    return Pola,Subjective
```

```python
for i in range(0,len(no_of_sent)):
  Scores=polarscore(POSITIVE_SCORE[i],NEGATIVE_SCORE[i],no_of_sent[i])
#print(POSITIVE_SCORE[0],NEGATIVE_SCORE[0],no_of_sent[0])
  print(Scores)
# break
  Polarityscore.append(Scores[0])
  Subjectivescore.append(Scores[1])
```
