```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
import time
```

```python
i=0
```

```python
Movies=[]
Runtime=[]
```

```python
while i <=140:
   no=np.arange(1001,8001,50)
   No=no[i]
   url=(f'https://www.imdb.com/search/title/?release_date=2018-01-01,2018-12-31&sort=num_votes,desc&start={No}&view=advanced')
   time.sleep(0.1)
   response=requests.get(url)
   soup=BeautifulSoup(response.text,'html.parser')
   title=soup.find_all('div',class_='lister-item-content')
   #title=soup.find_all('h3',class_='lister-item-header')
   for topic in title:
     #print(topic)
     name=topic.find('a').text
     #print(name)
     #print(name.text)

     if topic.find('small'):
        episode=topic.findAll('a')[1].text

        print(episode)

        name=name+' : '+episode
        Movies.append(name)
     else:
        Movies.append(name)
     if(topic.find('span',class_='runtime')):
        Runtime.append(topic.find('span',class_='runtime').text)
     else:
        Runtime.append(0)

   print(len(Movies))
   print(len(Runtime))
   i+=1
```
